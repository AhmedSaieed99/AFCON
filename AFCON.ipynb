# Africa Cup Of Nations

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Africa_Cup_of_Nations#Trophy_and_medals'
page = requests.get(url)
```

```python
data = BeautifulSoup(page.text, 'html')
```

```python
table = data.find_all('table')[4]
```

```python
columns = table.find_all('th')
winners = [title.text.strip() for title in columns]
print(winners)
```

```python
import pandas as pd
df = pd.DataFrame(columns = winners)
df
```

```python
column_data = table.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [info.text.strip() for info in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```
